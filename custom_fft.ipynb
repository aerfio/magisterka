```python
import tensorflow as tf
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np


x=np.linspace(0, 10*np.pi, 1000)
y=np.sin(2*np.pi*x)

df = pd.DataFrame(data={"y": y, "x": x})
df
```

```python
df["y"].plot(figsize=(15,8))
```

```python


signal="y"
fft = tf.signal.rfft(df[signal])
# fft = tf.signal.rfft(df["Day sin"])
f_per_dataset = np.arange(0, len(fft))

n_samples_h = len(df[signal])
hours_per_year = 24*365.2524
years_per_dataset = n_samples_h/(hours_per_year)

f_per_year = f_per_dataset/years_per_dataset

plt.step(f_per_year, np.abs(fft))
# plt.xscale('log')
plt.title("O")
# plt.ylim(0, 150)
plt.xlim([0.1, max(plt.xlim())])
# plt.xticks([1, 2*np.pi], labels=['1/Rok', '1/Dzień'])
_ = plt.xlabel('Częstotliwość (skala logarytmiczna)')
```
